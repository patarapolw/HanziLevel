```python
params = {
            'type': 'deck',
            'key': 'name'
        }
list(anki.getCardQuery('Chinese::Vocab', params))
```

```python
output = ''
for card in anki.getCardQuery('Chinese::Vocab', params):
    for char in card['note']['content'][0]:
        if char not in output:
            output += char
output
```
